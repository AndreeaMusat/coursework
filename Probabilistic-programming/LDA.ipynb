```python
import numpy as np
import matplotlib.pyplot as plt

from lda import LatentDirichletAllocation
from utils import select_data, text_to_idx

%reload_ext autoreload
%autoreload 2
%matplotlib inline
```

## Fetching the data

```python
from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups(
    data_home='./data', subset='train',
    categories=['comp.windows.x', 'rec.sport.hockey',
                'misc.forsale', 'talk.religion.misc'],
    remove=('headers', 'footers', 'quotes'),
    random_state=np.random.randint(1000),
    shuffle=False)

train_data, train_targets = select_data(data.data, data.target, 10)

print(len(train_data))
print(train_data[0])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vocab_size = 250
word_counter = CountVectorizer(
    lowercase=True, 
    stop_words='english',
    min_df=0.01,
    max_df=0.05,
    max_features=vocab_size,
    token_pattern='[a-zA-Z][a-zA-Z]+',
)

x_feat = word_counter.fit_transform(train_data)

print('The shape of the features is: {}'.format(x_feat.toarray().shape))
print('The selected features are: {}'.format(word_counter.vocabulary_))

idx_to_word = { idx : word for word, idx in word_counter.vocabulary_.items()}
train_data_num = text_to_idx(train_data, word_counter.vocabulary_)
```

## Latent Dirichlet Allocation

```python
lda = LatentDirichletAllocation(n_components=4, max_iter=500, burn_iter=100, n_jobs=2, vocab_size=vocab_size)
lda.fit(train_data_num)
```

```python
lda.plot_topics(idx_to_word)
```

```python
lda.plot_document_topics(train_data, lda.thetas)
```

## LDA used for computing the similarity between documents

LDA can be used to measure the similarity between documents by using the topic distribution for each document and computing the similarity between those distributions. For example, cosine similarity can be used:

If we have two vectors A and B and $\theta$ is the angle between them, then:

$\text{similarity} = cos(\theta) = \frac{A \dot B}{\left\lVert A\right\rVert \left\lVert B \right\rVert } $

We randomly choose 5 documents, print them and then plot the pairwise similarity matrix as an image.

```python
lda.plot_document_similarity(train_data, train_targets)
```

## Inferring the topic distribution for a new document

Looking at our fitted LDA model, we notice that $\theta$, $z$ and $w$ depend on our training data, while $\phi$ is fixed (we consider that once we learned the word distribution for each topic, it doesn't change anymore), so in order to infer the topic distribution for new documents, we need a new model in which $\phi$ has the value obtained in fit() (self.phis.value).

We fetch the data again, this time shuffled and select a few examples for each label. As the data is shuffled, the probability p of choosing a document that was already used for training is small:

$p = \frac{\text{num train documents}}{\text{num documents}} = \frac{40}{2155} = 0.0185 $

```python
data = fetch_20newsgroups(
    data_home='./data', subset='train',
    categories=['comp.windows.x', 'rec.sport.hockey',
                'misc.forsale', 'talk.religion.misc'],
    remove=('headers', 'footers', 'quotes'),
    random_state=np.random.randint(1000),
    shuffle=True)

test_data, test_targets = select_data(data.data, data.target, 5)
test_data_num = text_to_idx(test_data, word_counter.vocabulary_)
```

```python
test_thetas = lda.infer_topic_for_new_documents(test_data_num)
```

```python
lda.plot_document_topics(test_data, test_thetas)
```
